Imports

```python
import os
import datetime

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

Base

```python
df = pd.read_csv("nelson.csv", header=[0])
df = df.reset_index()
df.head(5)
```

```python
df['Fadiga'] = [1 if x > ((4/5) * len(df)) else 0 for x in df['index']]
```

```python
df.head(5)
```

```python
tam_janela = 50
split = 2
nova_df = []
janela = []


for i in df['Skeleton 016:LFTC Y']:
    janela.append(i)
    if (len(janela) == tam_janela):
        nova_df.append(janela)
        for i in range(split):
            janela.pop(0)
        continue
nova_df = np.array(nova_df)
nova_df[1]
```

Treino e Teste

```python
column_indices = {name: i for i, name in enumerate(df.columns)}
df = df.sample(frac = 1)
n = len(df)
train_df = df[0:int(n*0.7)]
val_df = df[int(n*0.7):int(n*0.9)]
test_df = df[int(n*0.9):]

train_x,train_y = df[['Skeleton 016:LFTC X','Skeleton 016:LFTC Y','Skeleton 016:LFTC Z']], df['Fadiga']
val_x,val_y = df[['Skeleton 016:LFTC X','Skeleton 016:LFTC Y','Skeleton 016:LFTC Z']], df['Fadiga']
test_x,test_y = df[['Skeleton 016:LFTC X','Skeleton 016:LFTC Y','Skeleton 016:LFTC Z']], df['Fadiga']
num_features = df.shape[1]
```

Normalização

```python
train_mean = train_x.mean()
train_std = train_x.std()

train_x = (train_x - train_mean) / train_std
val_x = (val_x - train_mean) / train_std
test_x = (test_x - train_mean) / train_std
```

```python
num_classes = len(np.unique(train_y))
```

```python
train_x = train_x.to_numpy()
val_x = val_x.to_numpy()
test_x = test_x.to_numpy()
```

```python
train_x = train_x.reshape((train_x.shape[0], train_x.shape[1], 1))
test_x = test_x.reshape((test_x.shape[0], test_x.shape[1], 1))
```

```python
train_x.shape
```

Modelo

```python
def make_model_lstm(input_shape):
    input_layer = tf.keras.layers.Input(input_shape)
    #embedding = tf.keras.layers.Embedding(input_dim=input_shape, output_dim=64)(input_layer)
    lstm = tf.keras.layers.LSTM(128)(input_layer)
    output_layer = tf.keras.layers.Dense(num_classes)(lstm)
    return tf.keras.models.Model(inputs=input_layer, outputs=output_layer)
```

```python
def make_model_rnn(input_shape):
    input_layer = tf.keras.layers.Input(input_shape)
    #embedding = tf.keras.layers.Embedding(input_dim=input_shape, output_dim=64)(input_layer)
    lstm = tf.keras.layers.SimpleRNN(128)(input_layer)
    output_layer = tf.keras.layers.Dense(num_classes)(lstm)
    return tf.keras.models.Model(inputs=input_layer, outputs=output_layer)
```

```python
def make_model_cnn(input_shape):
    input_layer = tf.keras.layers.Input(input_shape)
    #embedding = tf.keras.layers.Embedding(input_dim=input_shape, output_dim=64)(input_layer)
    cn2 = tf.keras.layers.Conv1D(32,kernel_size=3,padding="same")(input_layer)
    max = tf.keras.layers.MaxPooling1D()(cn2)
    flat = tf.keras.layers.Flatten()(max)
    output_layer = tf.keras.layers.Dense(num_classes)(flat)
    return tf.keras.models.Model(inputs=input_layer, outputs=output_layer)
```

```python
def make_model(input_shape):
    input_layer = tf.keras.layers.Input(input_shape)

    conv1 = tf.keras.layers.Conv1D(filters=64, kernel_size=3, padding="same")(input_layer)
    conv1 = tf.keras.layers.BatchNormalization()(conv1)
    conv1 = tf.keras.layers.ReLU()(conv1)

    conv2 = tf.keras.layers.Conv1D(filters=64, kernel_size=3, padding="same")(conv1)
    conv2 = tf.keras.layers.BatchNormalization()(conv2)
    conv2 = tf.keras.layers.ReLU()(conv2)

    conv3 = tf.keras.layers.Conv1D(filters=64, kernel_size=3, padding="same")(conv2)
    conv3 = tf.keras.layers.BatchNormalization()(conv3)
    conv3 = tf.keras.layers.ReLU()(conv3)

    gap = tf.keras.layers.GlobalAveragePooling1D()(conv3)

    output_layer = tf.keras.layers.Dense(num_classes, activation="softmax")(gap)

    return tf.keras.models.Model(inputs=input_layer, outputs=output_layer)
```

```python
model = make_model(input_shape= train_x.shape[1:])
#tf.keras.utils.plot_model(model, show_shapes=True)
model.summary()
```

```python
epochs = 500
batch_size = 32

callbacks = [
    tf.keras.callbacks.ModelCheckpoint(
        "best_model.keras", save_best_only=True, monitor="val_loss"
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.5, patience=10, min_lr=0.0001
    ),
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=20, verbose=1),
]
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["sparse_categorical_accuracy"],
)
history = model.fit(
    train_x,
    train_y,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=callbacks,
    validation_split=0.2,
    verbose=1,
)
```

```python
model = tf.keras.models.load_model("best_model.keras")

test_loss, test_acc = model.evaluate(test_x, test_y)

print("Test accuracy", test_acc)
print("Test loss", test_loss)
```

```python
metric = "sparse_categorical_accuracy"
plt.figure()
plt.plot(history.history[metric])
plt.plot(history.history["val_" + metric])
plt.title("model " + metric)
plt.ylabel(metric, fontsize="large")
plt.xlabel("epoch", fontsize="large")
plt.legend(["train", "val"], loc="best")
plt.show()
plt.close()
```
